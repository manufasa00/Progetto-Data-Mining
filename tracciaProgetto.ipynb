```python
columns_to_keep = ['Were_underweight', 'Were_overweight','Were_obese']
dataset = dataset.dropna(subset=[col for col in dataset.columns if col not in columns_to_keep])
```

```python
for col in dataset.columns:
    percent_missing_value = (dataset[col].isnull().sum()/dataset.shape[0]) * 100
    print("Percent of missing values for the column ", col, " is ", percent_missing_value)
```

Rinomino le colonne con nomi troppo lunghi

```python
new_cols = {'Bullied_on_school_property_in_past_12_months': 'Bullied_in_school',
           'Bullied_not_on_school_property_in_past_12_months': 'Bullied_outside_school',
           'Cyber_bullied_in_past_12_months': 'Cyber_bullied'}

dataset = dataset.rename(columns=new_cols)
```

```python
new_dataset = dataset.dropna(subset=['Were_underweight', 'Were_overweight', 'Were_obese'])
new_dataset.to_csv('new_dataset.csv', index=False)
```

Adesso effettuo una pulizia del dataset e controllo i valori per ogni colonna

```python
ord = {'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Most of the time': 3, 'Always': 4}
dataset.Felt_lonely = dataset.Felt_lonely.map(ord)
dataset.Other_students_kind_and_helpful = dataset.Other_students_kind_and_helpful.map(ord)
dataset.Parents_understand_problems = dataset.Parents_understand_problems.map(ord)

ord_2 = {"0 times": 0, "1 time": 1, "2 or 3 times" : 2, "4 or 5 times" : 3, "6 or 7 times" : 4, "8 or 9 times" : 5,
            "10 or 11 times" : 6, "12 or more times" : 7}
dataset.Physically_attacked = dataset.Physically_attacked.map(ord_2)
dataset.Physical_fighting = dataset.Physical_fighting.map(ord_2)

ord_3 = {"0": 0, "1": 1, "2" : 2, "3 or more" : 3}
dataset.Close_friends = dataset.Close_friends.map(ord_3)

ord_4 = {"0 days": 0, "1 or 2 days": 1, "3 to 5 days" : 2, "6 to 9 days" : 3, "10 or more days" : 4}
dataset.Miss_school_no_permission = dataset.Miss_school_no_permission.map(ord_4)

ord_5 = {'11 years old or younger' : 1, '12 years old' : 2, '13 years old' : 3, '14 years old' : 4, '15 years old' : 5,
       '16 years old' : 6, '17 years old' : 7, '18 years old or older' : 8}
dataset.Custom_Age = dataset.Custom_Age.map(ord_5)

columns = ['Felt_lonely', 'Other_students_kind_and_helpful', 'Parents_understand_problems', 'Physically_attacked',
           'Physical_fighting', 'Close_friends', 'Miss_school_no_permission', 'Custom_Age']

for col in columns:
    dataset[col] = dataset[col].astype(int)
```

Ora che sono stati rimossi i null si può fare la predizione per le 3 colonne una alla volta

```python
from sklearn.preprocessing import LabelEncoder

# encode categorical variables

categorical_columns = ['Bullied_in_school', 'Bullied_outside_school', 'Cyber_bullied', 'Sex', 'Most_of_the_time_or_always_felt_lonely',
                       'Missed_classes_or_school_without_permission']

le = LabelEncoder()

for col in categorical_columns:
    dataset[col] = le.fit_transform(dataset[col])
```

```python
for col in dataset.columns:
    print(dataset[col].value_counts(), '\n')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Split the dataset into two parts: one with non-null values and another with null values
train_data = dataset.dropna(subset=['Were_underweight', 'Were_overweight', 'Were_obese'])
test_data = dataset[dataset['Were_underweight'].isnull() | dataset['Were_overweight'].isnull() | dataset['Were_obese'].isnull()]

# Separate the features and target columns
X_train = train_data.drop(['Were_underweight', 'Were_overweight', 'Were_obese'], axis=1)
y_train = train_data[['Were_underweight', 'Were_overweight', 'Were_obese']]
X_test = test_data.drop(['Were_underweight', 'Were_overweight', 'Were_obese'], axis=1)

# Train a classification model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict the target values for the dataset with null values
y_pred = model.predict(X_test)

# Replace the missing values in the target columns with the predicted values
dataset.loc[test_data.index, ['Were_underweight', 'Were_overweight', 'Were_obese']] = y_pred
```

```python
def one_hot_encode(df, column):
    encoded_df = pd.get_dummies(df[column], prefix=column, dtype=int)
    df = pd.concat([df, encoded_df], axis=1)
    df = df.drop(column, axis=1)  # Remove the original column
    return df

columns_ordinal = ['Felt_lonely', 'Other_students_kind_and_helpful', 'Parents_understand_problems', 'Physically_attacked',
           'Physical_fighting', 'Close_friends', 'Miss_school_no_permission', 'Custom_Age']

for col in dataset.columns:
    if col not in columns_ordinal:
        dataset = one_hot_encode(dataset, col)
```

```python
# For instance, printing highly correlated pairs above a certain threshold:
threshold = 0.3
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > threshold and abs(correlation_matrix.iloc[i, j]) < 0.9999:
            high_corr_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))
print("Highly correlated feature pairs:")
for pair in high_corr_pairs:
    print(f"{pair[0]} - {pair[1]} : {pair[2]}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Select the most correlated features
selected_features = ['Bullied_in_school_1', 'Bullied_in_school_0', 'Bullied_outside_school_1', 'Bullied_outside_school_0', 'Cyber_bullied_1',
                     'Cyber_bullied_0', 'Most_of_the_time_or_always_felt_lonely_1', 'Most_of_the_time_or_always_felt_lonely_0',
                     'Felt_lonely', 'Missed_classes_or_school_without_permission_1', 'Missed_classes_or_school_without_permission_0',
                     'Miss_school_no_permission', 'Were_overweight_Yes', 'Were_overweight_No', 'Were_obese_Yes', 'Were_obese_No']

# Create a new dataframe with the selected features
selected_df = dataset[selected_features]

# Perform clustering using K-means

kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(selected_df)

# Add the cluster labels to the dataframe
selected_df['Cluster'] = clusters

# Print the dataframe with cluster labels
print(selected_df)
# Calculate the silhouette score
silhouette_avg = silhouette_score(selected_df, clusters)

# Print the silhouette score
print("The average silhouette score is:", silhouette_avg)
```

```python
distorsions = []
K = range(1, 30)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(dataset)
    distorsions.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distorsions, 'bx-')
plt.xlabel('Numero di Cluster')
plt.ylabel('Distorsione')
plt.title('Elbow Method')
plt.show()
```
