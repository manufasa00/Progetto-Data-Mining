```python
import pandas as pd
```

```python
df = pd.read_csv('new_dataset.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

for col in df.columns:
    if col not in ['Bullied_in_school', 'Bullied_outside_school',
                   'Cyber_bullied']:
        values = df.groupby(['Bullied_in_school', 'Bullied_outside_school',
                             'Cyber_bullied'])[col].get_group(('Yes', 'Yes', 'Yes')).value_counts(normalize=True) * 100
        values.plot(kind='bar')
        plt.title(f'Percentage of {col} for (Yes, Yes, Yes)')
        plt.xlabel(col)
        plt.ylabel('Percentage')
        plt.show()
                 
```
